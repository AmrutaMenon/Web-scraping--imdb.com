# Web Scraping from imdb.com

```python
#importing required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from IPython.display import clear_output
```

```python
#creating empty lists inorder to append
name=[]
yr=[]
certif=[]
stars=[]
duration=[]
genre=[]
votes=[]
metascore=[]

#base url of the site
base_url='https://www.imdb.com/search/title/?release_date=2021-01-01,2021-12-31&sort=num_votes,desc&start='

#scraping 550 movie info 
for i in range(1,502,50):
    
    time.sleep(3)
    print(f'Scrapping for {i}')
    clear_output(wait=True)
    
    url=base_url+str(i)+'&ref_=adv_nxt'
    req=requests.get(url)
    soup=BeautifulSoup(req.text,'html')
    
    
    #movie_name
    nam=soup.findAll('h3',attrs={'class':'lister-item-header'})
    for j in nam:
        name.append(j.a.text)
        
    
    #year_of_release
    yer=soup.findAll('h3',{'class':'lister-item-header'})
    for k in yer:
        yr.append((k.find('span',attrs={'class':'lister-item-year text-muted unbold'}).text.replace('(','').replace(')','')))
        
    
    #duration
    full=soup.findAll('div',{'class':'lister-item mode-advanced'})
    for l in full:
        t=l.find('span',{'class':'runtime'})
        if t is not None:
            duration.append(t.text)
        else:
            duration.append("Not available")
        
      
    #certification
    c=soup.findAll('div',{'class':'lister-item mode-advanced'})
    for m in c:
        certif.append(m.find('p',{'class':'text-muted'}).span.text)
        
    #star
    star=soup.findAll('div',attrs={'class':'ratings-bar'})
    for n in star:
        stars.append(n.strong.text)
        
    #metascore
    mta=soup.findAll('div',{'class':'ratings-bar'})
    for o in mta:
        a=o.find('span',attrs={'class':'metascore'})
        if a is not None:
            metascore.append(a.text.strip())
        else:
            metascore.append('Not available')
            
    #genre
    genr=soup.findAll('span',attrs={'class':'genre'})
    for p in genr:
        genre.append(p.text.strip())
        
    #votes
    v=soup.findAll('p',{'class':'sort-num_votes-visible'})
    for q in v:
        votes.append(q.find('span',{'name':'nv'}).text)

```

```python
#creating data frame

data=pd.DataFrame({'Movie_name':name,
                  'Year_of_release':yr,
                  'Certification':certif,
                  'Duration':duration,
                  'imdb_rating':stars,
                  'Genre':genre,
                  'Votes':votes,
                  'Metascore':metascore})
data
```

```python
data.to_csv('imdb_scraped_data.csv',index=False)
```
